```python





# Embed Knowledge Base
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
texts = [
    f"{item['category']} {item['summary']} {' '.join(item['key_points'])}"
    for item in knowledge_base
]
embeddings = np.array(embedding_model.encode(texts))

# Build FAISS Index
dimension = embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(dimension)
faiss_index.add(embeddings)

# Define Retrieval Function
def retrieve_suggestions(category, top_k=3):
    query_embedding = np.array(embedding_model.encode([category]))
    distances, indices = faiss_index.search(query_embedding, top_k)
    suggestions = [{"category": knowledge_base[idx]["category"],
                    "summary": knowledge_base[idx]["summary"],
                    "key_points": knowledge_base[idx]["key_points"]}
                   for idx in indices[0]]
    return suggestions

# Define RAG Pipeline
def rag_response(user_query, spending_category):
    # Retrieve relevant suggestions
    suggestions = retrieve_suggestions(spending_category, top_k=1)
    retrieved_text = f"Category: {suggestions[0]['category']}\nSummary: {suggestions[0]['summary']}\nKey Points: {'; '.join(suggestions[0]['key_points'])}"

    # Combine user query with retrieved knowledge
    prompt = (
        f"User query: {user_query}\n"
        f"Retrieved information:\n{retrieved_text}\n"
        f"Provide personalized advice based on the user's query and the retrieved information."
    )

    # Generate response
    response = llama_pipe(prompt, max_length=400, num_return_sequences=1)
    return response[0]["generated_text"]

# Example Usage
user_query = "I'm overspending on dining out. What can I do?"
spending_category = "Dining Out"

response = rag_response(user_query, spending_category)
print(response)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import numpy as np

# Example training data (queries and their categories)
# Updated training queries and categories
training_queries = [
    "How can I reduce my rent expenses?",
    "What's the best way to find shared accommodations?",
    "How do I negotiate lease terms with landlords?",
    "Tips for saving money on groceries.",
    "How do I use cashback apps effectively?",
    "What are the benefits of buying in bulk?",
    "Is public transport cheaper than driving?",
    "How can I save on fuel costs?",
    "What are the benefits of cycling to work?",
    "How to lower my electricity bill?",
    "What are energy-saving tips for winter?",
    "Where can I find free events in my city?",
    "How to save on subscriptions and movies?"
]

categories = [
    "Accommodation", "Accommodation", "Accommodation",
    "Groceries", "Groceries", "Groceries",
    "Transport", "Transport", "Transport",
    "Utilities", "Utilities",
    "Entertainment", "Entertainment"
]

# Vectorize text data and train the classifier
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(training_queries)
classifier = LogisticRegression()
classifier.fit(X_train, categories)

# Vectorize text data and train a simple classifier


def predict_category(user_query):
    query_vector = vectorizer.transform([user_query])
    predicted_category = classifier.predict(query_vector)[0]
    return f"The query is related to: {predicted_category}"
```

```python
# Average monthly expenses for Hyderabad
average_expenses = {
    "Bachelors": {
        "Accommodation": (4000, 9500),
        "Groceries": (1000, 2500),
        "Transport": (1000, 2000),
        "Utilities": (1200, 2500),
        "Entertainment": (4000, 7000),
        "Shopping": (4000, 5000),
        "Kids Schooling": (0, 0)
    },
    "Couples": {
        "Accommodation": (12000, 28000),
        "Groceries": (2000, 5500),
        "Transport": (1500, 3500),
        "Utilities": (2000, 3500),
        "Entertainment": (2000, 7000),
        "Shopping": (3000, 5500),
        "Kids Schooling": (0, 0)
    },
    "Families": {
        "Accommodation": (12000, 38000),
        "Groceries": (2500, 11000),
        "Transport": (1500, 4000),
        "Utilities": (2000, 3500),
        "Entertainment": (2000, 8000),
        "Shopping": (3500, 6500),
        "Kids Schooling": (3000, 18000)
    }
}

def calculate_average(expenses_range):
    """Calculate the average expense in a given range."""
    return sum(expenses_range) / 2

def compare_expenses(user_type, user_expenses):
    """Compare user's expenses with average values and identify overspending."""
    results = {}
    total_avg = 0
    total_user = 0

    for category, user_spending in user_expenses.items():
        if category in average_expenses[user_type]:
            avg_spending = calculate_average(average_expenses[user_type][category])
            total_avg += avg_spending
            total_user += user_spending

            if user_spending > avg_spending:
                results[category] = {
                    "User Spending": user_spending,
                    "Average Spending": avg_spending,
                    "Overspending": user_spending - avg_spending
                }

    # Calculate the total spending comparison
    total_comparison = {
        "Total User Spending": total_user,
        "Total Average Spending": total_avg,
        "Overspending": max(0, total_user - total_avg)
    }

    return results, total_comparison

# Sample user expenses data (example for a family)
user_type = "Families"
user_expenses = {
    "Accommodation": 25000,
    "Groceries": 8000,
    "Transport": 3000,
    "Utilities": 3000,
    "Entertainment": 9000,
    "Shopping": 6000,
    "Kids Schooling": 15000
}

# Perform the expense comparison
over_spending_details, total_comparison = compare_expenses(user_type, user_expenses)

# Output the comparison results
print("Detailed Overspending Analysis:")
for category, details in over_spending_details.items():
    print(f"{category}: User spent {details['User Spending']} compared to average {details['Average Spending']} "
          f"(Overspending by {details['Overspending']})")

print("\nTotal Monthly Expense Comparison:")
print(f"User's total monthly spending: {total_comparison['Total User Spending']}")
print(f"Average monthly spending: {total_comparison['Total Average Spending']}")
print(f"Total overspending: {total_comparison['Overspending']}")

# This output can be fed into the LLM to generate advice on cost-saving.
```

```python
def rag_response_based_on_expenses(user_type, user_expenses):
    # First, calculate overspending details for each category
    over_spending_details, total_comparison = compare_expenses(user_type, user_expenses)

    overspending_categories = []  # List to hold categories with overspending

    # Identify which categories have overspending
    for category, details in over_spending_details.items():
        if details["Overspending"] > 0:
            overspending_categories.append(category)

    # If there are overspending categories, retrieve advice
    if overspending_categories:
        for category in overspending_categories:
            user_query = f"I'm overspending on {category}. What can I do?"
            response = rag_response(user_query)
            print(f"Advice for {category} overspending: {response}")
    else:
        print("You are not overspending in any of the categories.")

# Example usage
user_type = "Families"  # Assuming the user is a family
user_expenses = {
    "Accommodation": 25000,
    "Groceries": 8000,
    "Transport": 3000,
    "Utilities": 3000,
    "Entertainment": 9000,
    "Shopping": 6000,
    "Kids Schooling": 15000
}

# Perform the expense comparison and generate advice based on overspending
rag_response_based_on_expenses(user_type, user_expenses)
```
